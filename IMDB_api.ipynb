```python
import pandas as pd
import requests
import json
import numpy as np
apikey = ''
url = "http://www.omdbapi.com/?apikey=" + apikey + "&t="
```

```python
movies_df = pd.read_csv("../ETL-PROJECT/Resources/netflix_movies.csv")
```

```python
movies = movies_df['title']

responses = []

for movie in movies:
    movie_data = requests.get(url + movie).json()
    responses.append(movie_data)
```

```python
print(responses)
```

```python
with open("OMDB.json", "w") as outfile:
    json.dump(responses, outfile, indent=4)
```
