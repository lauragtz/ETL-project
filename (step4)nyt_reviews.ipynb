```python
# Dependencies
import pandas as pd
import requests
from config import api_key
from config import password
import time
import json
from sqlalchemy import create_engine
```

```python
#open connection to Postgress
connection_string = f"postgres:{password}@localhost:5432/Netflix_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check to make sure there is a connection 
movies_df = pd.read_sql_query('select * from netflix_movies', con=engine)
movies_df.head()
```

```python
#setup URL for API pull
url = "https://api.nytimes.com/svc/movies/v2/reviews/search.json?"

#setup dataframe for adding json data
movie_reviews_df = pd.DataFrame()

#loop to go through ever movie in the data base and add the result to a dataframe
for movie in movies_df['title']:
    print(movie)
    movie = movie.replace(" ", "+")
    query = movie
    try:
        query_url = f"{url}api-key={api_key}&query={query}"
        movie_result = requests.get(query_url).json()
        movie_result = pd.json_normalize(movie_result, 'results')
        movie_reviews_df = movie_reviews_df.append(movie_result)
    except:
        pass

movie_reviews_df
```

```python
#clean up the results
movie_reviews_df.drop(columns=['multimedia.type', 'multimedia.src', 'multimedia.width','multimedia.height'], inplace=True)
movie_reviews_df.rename(columns={'link.type': 'link_type', 'link.url': 'link_url', 'link.suggested_link_text': 'link_suggested_link_text' }, inplace=True)
movie_reviews_df.columns
```

```python
#send data to postgres database
movie_reviews_df.to_sql(name='ny_times_reviews', con=engine, if_exists='append', index=False)
```
